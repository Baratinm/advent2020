```python
%reset -f
import string

class Person():
    def __init__(self):
        self.ecl=""
        self.byr=""
        self.iyr=""
        self.eyr=""
        self.hgt=""
        self.hcl=""
        self.pid=""
        self.cid=""
    #cid is optional
    def is_good(self):
        if (self.ecl!="" and
            self.byr!="" and
            self.iyr!="" and
            self.eyr!="" and
            self.hgt!="" and
            self.hcl!="" and
            self.pid!=""):
            return 1
        else:
            return 0
    
    def good_ecl(self):
        ecl= self.ecl
        if(ecl in "amb blu brn gry grn hzl oth" and len(ecl) == 3):
            return True
        else:
            print("bad ecl : " + str(ecl))
            return False
    def good_byr(self):
        byr = self.byr
        if(len(byr)==4 and int(byr)>=1920 and int(byr)<=2002):
            return True
        else:
            print("bad byr : " + str(byr))
            return False
        
    def good_iyr(self):
        iyr = self.iyr
        if(len(iyr)==4 and int(iyr)>=2010 and int(iyr)<=2020):
            return True
        else:
            print("bad iyr : " + str(iyr))
            return False
        
    def good_eyr(self):
        eyr = self.eyr
        if(len(eyr)==4 and int(eyr)>=2020 and int(eyr)<=2030):
            return True
        else:
            print("bad eyr : " + str(eyr))
            return False
        
    def good_hgt(self):
        hgt = self.hgt
        if("cm" in hgt and 150 <= int(hgt[:-2]) <= 193):
            return True
        elif("in" in hgt and 59 <= int(hgt[:-2]) <= 76):
            return True
        else:
            print("bad hgt : " + str(hgt))
            return False
    
    def good_hcl(self):
        hcl = self.hcl
        if(len(hcl) == 7 and hcl[0] == "#" and all(c in string.hexdigits for c in hcl[1:])):
            return True
        else:
            print("bad hcl : " + str(hcl))
            return False

    def good_pid(self):
        pid = self.pid
        if(pid.isdigit() and len(pid)==9):
            return True
        else:
            print("bad pid : " + str(pid))
            return False
    
    
    def is_gooder(self):
        if(self.good_ecl() and self.good_byr() and self.good_iyr() and self.good_eyr() and self.good_hgt() and self.good_hcl() and self.good_pid()):
            return True
        else:
            return False

    
    def display(self):
        print("ecl = " + self.ecl + ", byr = " + self.byr + ", iyr = " + self.iyr + ", eyr = " + self.eyr + ", hgt = " + self.hgt + ", hcl = " + self.hcl + ", pid = " + self.pid + ", cid = " + self.cid)           
   
"""
ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
byr (Birth Year) - four digits; at least 1920 and at most 2002.
iyr (Issue Year) - four digits; at least 2010 and at most 2020.
eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
hgt (Height) - a number followed by either cm or in:
If cm, the number must be at least 150 and at most 193.
If in, the number must be at least 59 and at most 76.
hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
pid (Passport ID) - a nine-digit number, including leading zeroes.
cid (Country ID) - ignored, missing or not.
"""
    
       
p = Person()
valid = 0
more_valid = 0
with open("Input_Day_4.txt", "r") as a_file:
    for line in a_file:
        stripped_line = line.strip()
        if(stripped_line==''):
            if p.is_gooder():
                    more_valid+=1
                    print("AAAAAAAAAAAAAAAAAAAAAAAAAAA")
            if(p.is_good()==1):
                valid+=1
                print("BBBBBBBBBBBBBBBBBBBBBBBBBBB")        
                p.display()
            else:
                print("CCCCCCCCCCCCCCCCCCCCCCCCCCC")
                p.display()
            p=Person()
            print("-------------------------")
        else:
            attributes = stripped_line.split()
            for att in attributes:
                if ("ecl:" in att):
                    p.ecl = att[4:]
                if ("byr:" in att):
                    p.byr = att[4:]
                if ("iyr:" in att):
                    p.iyr = att[4:]
                if ("eyr:" in att):
                    p.eyr = att[4:]
                if ("hgt:" in att):
                    p.hgt = att[4:]
                if ("hcl:" in att):
                    p.hcl = att[4:]
                if ("pid:" in att):
                    p.pid = att[4:]
                if ("cid:" in att):
                    p.cid = att[4:]

if(p.is_good()==1):
    valid+=1
    print("NEW VALID PERSON")
    p.display()
else:
    print("CCCCCCCCCCCCCCCCCCCCCCCCCCCC")
    p.display()
                    
print("There are " + str(valid) + " valid people and " + str(more_valid) + " super valid people")
        
        
```
